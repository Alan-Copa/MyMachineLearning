```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression

# Generate synthetic 2D data (as an example)
np.random.seed(0)
X = np.random.randn(100, 2)
y = (X[:, 0] ** 2 + X[:, 1] ** 2 < 1).astype(int)  # Circular boundary

# Add polynomial features (quadratic)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit logistic regression to transformed data
log_reg = LogisticRegression()
log_reg.fit(X_poly, y)

# Plot the data points and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = log_reg.predict(poly.transform(np.c_[xx.ravel(), yy.ravel()]))
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0.5], cmap="Greys", vmin=0, vmax=0.6)
plt.title("Polynomial Features (Quadratic) with Logistic Regression")
plt.show()
```

```python
from sklearn.svm import SVC

# Generate synthetic 2D data (same as before)
np.random.seed(0)
X = np.random.randn(100, 2)
y = (X[:, 0] ** 2 + X[:, 1] ** 2 < 1).astype(int)  # Circular boundary

# Fit SVM with RBF kernel
svm_rbf = SVC(kernel='rbf', gamma=1)
svm_rbf.fit(X, y)

# Plot the data points and decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = svm_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[0.5], cmap="Greys", vmin=0, vmax=0.6)
plt.title("RBF Kernel with SVM")
plt.show()
```
