```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

# Generating synthetic data
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # 100 points between 0 and 10
y = 2 * np.sin(X) + 1 + np.random.randn(100, 1)  # y = 2sin(x) + 1 + noise

# Creating polynomial features (e.g., X, X^2, X^3)
poly = PolynomialFeatures(degree=5)  # Try degree 5 for flexibility
X_poly = poly.fit_transform(X)

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Applying Ridge Regression with polynomial features
ridge = Ridge(alpha=1.0)  # Regularization strength (lambda = 1.0)
ridge.fit(X_train, y_train)

# Predicting and plotting
X_range = np.linspace(0, 10, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred = ridge.predict(X_range_poly)

plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_range, y_pred, color='red', label='Ridge Regression Fit')
plt.title('Ridge Regression with Polynomial Features')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Display the model coefficients
print(f"Ridge Coefficients: {ridge.coef_}, Intercept: {ridge.intercept_}")
```
